```python
import os


df_final = None 
file = os.getcwd()  # Get the current working directory
for files in os.listdir(file):
    if files.endswith('_filtered.csv'):
        df = pd.read_csv(os.path.join(file, files))
        if df_final is None:
            df_final = df
        else:
            df_final = pd.concat([df_final, df], ignore_index=True)
```

```python
non_tech = pd.read_csv('all_non_tech_jobs.csv')

non_tech_filtered = non_tech[["description" , "company_description"]]
non_tech_filtered.shape
```

```python
t1 = non_tech_filtered[non_tech_filtered['description'].notna() & non_tech_filtered['company_description'].notna()]
t1.shape
```

```python
t1.isna().sum()
```

```python
tech = pd.read_csv('all_tech_jobs.csv')

tech_filtered = tech[["description" , "company_description"]]
tech_filtered.shape
```

```python
t2 = tech_filtered[tech_filtered['description'].notna() & tech_filtered['company_description'].notna()]
t2.shape
```

```python
smtg = pd.read_csv('jobspy.csv')

smtg_filtered = smtg[["description" , "company_description"]]
smtg_filtered.shape
```

```python
t3 = smtg_filtered[smtg_filtered['description'].notna() & smtg_filtered['company_description'].notna()]
t3.shape
```

```python
merged = pd.concat([t1, t2, t3], ignore_index=True)
merged.shape
```

```python
merged.isna().sum()
```

```python
merged_duplicate = merged.drop_duplicates()
merged_duplicate.shape
```

```python
merged_duplicate.iloc[0,0] == merged_duplicate.iloc[1,0]
```

```python
merged_duplicate.to_csv('final_filtered_jobs.csv', index=False)
```
