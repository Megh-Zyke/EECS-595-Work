```python
import pandas as pd

# -----------------------------
# Step 1: Load CSVs
# -----------------------------
df_claude = pd.read_csv('/Users/sruthishivaramakrishnan/Documents/CSE595_project/outputs/claude_personas.csv')
df_gpt = pd.read_csv('/Users/sruthishivaramakrishnan/Documents/CSE595_project/outputs/GPT_personas.csv')
df_qwen = pd.read_csv('/Users/sruthishivaramakrishnan/Documents/CSE595_project/outputs/qwen_final_complete.csv')
df_llama = pd.read_csv('/Users/sruthishivaramakrishnan/Documents/CSE595_project/outputs/Llama_personas.csv')

# -----------------------------
# Step 2: Select relevant columns and rename
# -----------------------------
df_claude = df_claude[['description', 'ideal_candidate']].rename(columns={'ideal_candidate': 'Claude_Persona'})
df_gpt = df_gpt[['job_description', 'persona']].rename(columns={'persona': 'GPT_Persona', 'job_description': 'description'})
df_qwen = df_qwen[['description', 'Qwen_Persona']]
df_llama = df_llama[['job_description', 'optimal_candidate']].rename(columns={'optimal_candidate': 'Llama_Persona', 'job_description': 'description'})

# -----------------------------
# Step 3: Clean descriptions
# -----------------------------


# -----------------------------
# Step 4: Merge dataframes on cleaned description
# -----------------------------
df_merged = df_claude.merge(df_gpt, on='description', how='left')
df_merged = df_merged.merge(df_qwen, on='description', how='left')
df_merged = df_merged.merge(df_llama, on='description', how='left')

# -----------------------------
# Step 5: Check for any missing personas
# -----------------------------
missing = df_merged[df_merged[['GPT_Persona','Qwen_Persona','Llama_Persona']].isna().any(axis=1)]

print(f"Number of rows with missing personas: {len(missing)}")
print("Sample missing rows:\n", missing.head())

# -----------------------------
# Step 6: Save merged dataframe
# -----------------------------
df_merged = df_merged.dropna(subset=['GPT_Persona', 'Qwen_Persona', 'Llama_Persona'])
missing = df_merged[df_merged[['GPT_Persona','Qwen_Persona','Llama_Persona']].isna().any(axis=1)]

print(f"Number of rows with missing personas: {len(missing)}")
print("Sample missing rows:\n", missing.head())
# df_merged[[ 'bias_age', 'bias_gender',
#        'bias_lgbtq+', 'bias_family_status', 'bias_race_ethnic',
#        'bias_disability', 'bias_socioeconomic', 'bias_appearance',
#        'bias_proximity', 'bias_affinity', 'bias_education', 'biases_found']] = df_claude[['bias_age', 'bias_gender',
#        'bias_lgbtq+', 'bias_family_status', 'bias_race_ethnic','bias_disability', 'bias_socioeconomic', 'bias_appearance',
#        'bias_proximity', 'bias_affinity', 'bias_education', 'biases_found']]
df_merged[0:5250].to_csv('/Users/sruthishivaramakrishnan/Documents/CSE595_project/outputs/merged_personas_clean.csv', index=False)
print("Merged CSV with cleaned descriptions saved successfully!")
```
